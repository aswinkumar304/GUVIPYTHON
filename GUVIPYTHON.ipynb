```python
import tkinter as tk
import mysql.connector
def Create_Database():
    connect_database = mysql.connector.connect(
        host = "localhost",
        user = "root",
        password = "Tamilan@24"
    )
    cursor = connect_database.cursor()
    cursor.execute("CREATE DATABASE IF NOT EXISTS Accounts ")
    connect_database.commit()
    connect_database.close()
    
def Transaction_details():
    DateTime = DateTime_entry.get()
    TransactionID = TransactionID_entry.get()
    Name = Name_entry.get()
    Amount = Amount_entry.get()
    Remarks = Remarks_entry.get()
    connect_database = mysql.connector.connect(
        host = "localhost",
        user = "root",
        password = "Tamilan@24",
        database = "accounts"
    )
    cursor = connect_database.cursor()
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS transactiondetails(
    Date_Time VARCHAR(200),
    TransactionID VARCHAR(200),
    Name VARCHAR(200),
    Amount VARCHAR(200),
    Remarks VARCHAR(200)
    )
    """)
    cursor.execute("INSERT INTO transactiondetails(DateTime,TransactionID,Name,Amount,Remarks) VALUES(%s,%s,%s,%s,%s)",(DateTime,TransactionID,Name,Amount,Remarks))
    connect_database.commit()
    connect_database.close()
    DateTime_entry.delete(0,tk.END)
    TransactionID_entry.delete(0,tk.END)
    Name_entry.delete(0,tk.END)
    Amount_entry.delete(0,tk.END)
    Remarks_entry.delete(0,tk.END)
def retrieve_Transaction_details():
    connect_database = mysql.connector.connect(
        host = "localhost",
        user = "root",
        password = "Tamilan@24",
        database = "accounts"
    )
    cursor = connect_database.cursor()
    cursor.execute("SELECT * FROM transactiondetails")
    data = cursor.fetchall()
    connect_database.close()
    retrieve_box.delete(0,tk.END)
    for item in data:
        retrieve_box.insert(tk.END,f"Date_Time:{item[0]},TransactionID:{item[1]},Name:{item[2]},Amount:{item[3]},Remarks:{item[4]}")
Create_Database()
root = tk.Tk()
root.title("Transaction_details")

DateTime_label = tk.Label(root,text="Date_Time")
TransactionID_label = tk.Label(root,text="TransactionID")
Name_label = tk.Label(root,text="Name")
Amount_label = tk.Label(root,text="Amount")
Remarks_label = tk.Label(root,text="Remarks")

DateTime_entry = tk.Entry(root)
TransactionID_entry = tk.Entry(root)
Name_entry = tk.Entry(root)    
Amount_entry = tk.Entry(root)
Remarks_entry = tk.Entry(root)

Add_database_button = tk.Button(root,text="ADD TO DATABASE",command = Transaction_details )
retrieve_box = tk.Listbox(root,height=10,width=80)
retrieve_database_button = tk.Button(root, text = "RETRIEVE FROM DATABASE", command = retrieve_Transaction_details)
DateTime_label.grid(row = 0,column = 0)
TransactionID_label.grid(row = 1,column = 0)
Name_label.grid(row = 2,column = 0)
Amount_label.grid(row = 3,column = 0)
Remarks_label.grid(row = 4,column = 0)

DateTime_entry.grid(row = 0,column = 1)
TransactionID_entry.grid(row = 1,column = 1)
Name_entry.grid(row = 2,column = 1)    
Amount_entry.grid(row = 3,column = 1)
Remarks_entry.grid(row = 4,column = 1)

Add_database_button.grid(row = 5,columnspan = 2)
retrieve_box.grid(row = 6,columnspan = 2)
retrieve_database_button.grid(row = 7,columnspan = 2) 
root.mainloop()
```
